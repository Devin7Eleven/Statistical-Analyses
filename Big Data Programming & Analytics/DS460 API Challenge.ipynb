```
# Libraries
import requests
import json
import pyspark.sql.functions as F
from pyspark.sql.functions import *
```

```
# Test to make sure API works
api_key = 'd4699447783c424718374c68a6ae3af8'
lat = '33.451690'
lon = '-111.916550'
cnt = '16'
url = f"https://api.openweathermap.org/data/2.5/forecast/daily?lat=%s&lon=%s&cnt=%s&appid=%s&units=imperial" % (lat, lon, cnt, api_key)

response = requests.get(url)
responseJson = response.json()

# Turns JSON data into sparks dataframe RDD
data = spark.read.json(sc.parallelize([responseJson]))

display(data)
```

```
# Converts city column into tabular format
city = data.select(F.col('city.id').alias('id'),
                  F.col('city.coord.lat').alias('lat'),
                  F.col('city.coord.lon').alias('lon'),
                  F.col('city.timezone').alias('timezone'),
                  F.col('city.name').alias('name'),
                  F.col('city.population').alias('population'))

display(city)
```

```
# Select the 'list' column from the original DataFrame
listColumn = data.select(F.col('list'))

display(listColumn)
```

```
from pyspark.sql.functions import explode

# Explode the 'list' column to create a new row for each date
listExploded = listColumn.select(explode(F.col('list')).alias('listExploded'))

display(listExploded)
```

```
# Extract the desired columns from the exploded 'list' column, and convert timestamps to human-readable format
columns = listExploded.select(
    F.from_unixtime(F.col('listExploded.dt'), 'yyyy-MM-dd HH:mm:ss').alias('date'),
    F.col('listExploded.speed').alias('speed'),
    F.col('listExploded.gust').alias('gust'),
    F.col('listExploded.deg').alias('degree'),
    F.col('listExploded.clouds').alias('clouds'),
    F.col('listExploded.pressure').alias('pressure'),
    F.col('listExploded.humidity').alias('humidity'),
    F.col('listExploded.temp.min').alias('temp_min'),
    F.col('listExploded.temp.max').alias('temp_max'),
    F.col('listExploded.temp.day').alias('temp'),
    F.col('listExploded.feels_like').alias('feelsLike'),
    F.col('listExploded.rain').alias('rain'),
    F.struct(
        F.col('listExploded.weather.id').alias('id'),
        F.col('listExploded.weather.description').alias('description'),
        F.col('listExploded.weather.icon').alias('icon'),
        F.col('listExploded.weather.main').alias('main')
    ).alias('weather'),
    F.from_unixtime(F.col('listExploded.sunrise'), 'yyyy-MM-dd HH:mm:ss').alias('sunrise'),
    F.from_unixtime(F.col('listExploded.sunset'), 'yyyy-MM-dd HH:mm:ss').alias('sunset'),
    F.col('listExploded.pop').alias('pop'),
    F.col('listExploded.temp.morn').alias('temp_morn'),
    F.col('listExploded.temp.eve').alias('temp_eve'),
    F.col('listExploded.temp.night').alias('temp_night')
)

# Show the resulting DataFrame with each date as a separate row
display(columns)
```

```
# Combine the data
forecast = city.crossJoin(columns)

display(forecast)
```

# REST and GraphQL API Differences

The first major difference between REST and GraphQL is that while GraphQL uses a single API to call and retrieve data, REST utilizes multiple API calls. GraphQL, made by Facebook, was designed for executing queries with existing data. It's a method of fetching and updating data. REST, however, is a software architectural style that was designed for working with media components, files, or hardware devices with the purpose of creating web services. GraphQL extracts exactly what the user wants because the code allows for the user to specify exactly what they want into also directly into their own workspace. While made for simple queries, it is less prone to error and has auto validation for checking.

REST API designed for complex queries has some noticeable issues in that it needs to be conssitently added with more endpoints to the API call, leading to more possible coding errors and typos. There is also the issue with over or under fetching, or in other words, getting too much or too little data from the fetched soruce. REST uses endpoints to pull data in but they come in seperately, whereas GraphQL combines all fields into one table. Because of their difference in purpose, there is some overlap but also several differences to suit those needs. Both return JSON data in the request from a fetched URL but where in REST the endpoint you call is the identity of that object, in GraphQL the identity is separate from how you fetch it. In the end, GraphQL can be thought of a system for calling many nested endpoints in one request like a multiplexed REST.
