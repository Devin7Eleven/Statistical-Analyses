```
patterns = spark.table("safegraph.patterns")

lds_patterns = patterns.join(lds_church.select("placekey", "wkt_area_sq_meters", "street_address", "latitude", "longitude"), on = ['placekey'], how = "inner")
print(lds_patterns.count())
display(lds_patterns)
```

```
WindowSpec_order = Window.partitionBy("placekey").orderBy("mean_day")
WindowSpec = Window.partitionBy("placekey")

lds_day = lds_patterns\
    .withColumn("year", F.year("date_range_start"))\
    .filter(F.col("year") == 2019)\
    .select("placekey", "raw_visit_counts", "normalized_visits_by_state_scaling", F.explode("popularity_by_day"))\
    .withColumn("percent", F.col("value")/F.col("raw_visit_counts"))\
    .withColumn("normalized", F.col("normalized_visits_by_state_scaling") * F.col("percent"))\
    .withColumn("normalized_day", F.col("normalized")/4.3)\
    .groupby(["placekey", "key"])\
    .agg(
        F.round(F.mean("normalized_day")).alias("mean_day")
    )\
    .withColumn("isLargest", F.col("mean_day") == F.max("mean_day").over(WindowSpec))\
    .withColumn("rank", F.rank().over(WindowSpec_order))\
    .filter(F.col("rank").isin([6,7]))\
    .filter(F.col("key") == "Sunday")\
    .filter(F.col("mean_day") > 10)


# Maybe I should keep them all. Ignore this work. There should be more than 209 church buildings in Idaho.
display(lds_day)
```
