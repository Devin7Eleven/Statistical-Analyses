```
# Libraries
import pandas as pd
import pyspark.sql.functions as F
from pyspark.sql.functions import udf
from pyspark.sql.functions import avg
from pyspark.sql.types import StringType
from pyspark.sql.functions import when
from pyspark.sql.window import Window

# Data
poi = spark.table("safegraph.places") # Load in SafeGraph places
patterns = spark.table("safegraph.patterns") # Load in SafeGraph patterns
```

```
# Filters places dataset to just Ice Cream shops in Idaho
frozenDessert = poi.filter(
    (F.col("top_category") == "Restaurants and Other Eating Places") &
    (F.col("location_name").rlike("Ice|Cream|Frozen|Yogurt|Heladeria|Dairy|kiwi|Sundaes|Custard|Gelato|\b[Ii]ce [Cc]ream\b")) |
    (F.col("top_category") == "Restaurants and Other Eating Places") &
    (F.col("category_tags").rlike("Ice Cream Shop"))
)

display(frozenDessert)
```

```
# Return matching rows from patterns dataset
iceCreamPatterns = patterns.join(frozenDessert, on = ['placekey'], how = 'inner')

display(iceCreamPatterns)
```

```
def get_season(month):
  if month in [12,1,2]:
    return "Winter"
  elif month in [3,4,5]:
    return "Spring"
  elif month in [6,7,8]:
    return "Summer"
  else:
    return "Fall"

udfSeasons = udf(get_season, StringType())

# Below code doesn't work, get "cannot convert column into bool"
# iceCream = iceCreamData\
#     .withColumn('year', F.year('date_range_start'))\
#     .withColumn('month', F.month('date_range_start'))\
#     .withColumn('season', when(iceCream.month in [12, 1, 2], 'winter')
#                 .when(iceCream.month in [3, 4, 5], 'spring')
#                 .when(iceCream.month in [6, 7, 8], 'summer')
#                 .otherwise('fall'))

iceCream1 = iceCreamPatterns.select('placekey', 'date_range_start', 'Raw_visit_counts', 'raw_visitor_counts', 'distance_from_home', 'median_dwell', 'location_name')\
    .na.drop()\
    .dropDuplicates()\
    .withColumn('year', F.year('date_range_start'))\
    .withColumn('month', F.month('date_range_start'))\
    .withColumn('distanceFromHomeMiles', F.round(F.col('distance_from_home') / 1609.344, 2))

iceCream2 = iceCreamPatterns.select('placekey', 'date_range_start', 'Raw_visit_counts', 'raw_visitor_counts', 'distance_from_home', 'median_dwell', 'location_name')\
    .na.drop()\
    .dropDuplicates()\
    .withColumn('year', F.year('date_range_start'))\
    .withColumn('month', F.month('date_range_start'))\
    .withColumn('distanceFromHomeMiles', F.round(F.col('distance_from_home') / 1609.344, 2))\
    .agg(avg("distanceFromHomeMiles").alias("avg_dist"))

iceCream3 = iceCream1.join(iceCream2)\
    .withColumn('distance_var', F.round(F.col('distanceFromHomeMiles') - F.col('avg_dist'), 2))

display(iceCream3)
```
